```python

%%writefile ayeapp.py
import streamlit as st
import tensorflow.keras
from PIL import Image, ImageOps
import numpy as np



st.title('Eye diseases classification project ')
st.write("by STEM Students")
st.image("/content/drive/MyDrive/ayes project/tom-s-znUZEotYJv0-unsplash.jpg", width=600, height=500)
st.write('This project was made by STEM students to help people detect eye diseses')

image_input = st.file_uploader(label='Upload eye image (PNG, JPG or JPEG)', type=['png', 'jpg', 'jpeg'])
if image_input:
  img = image_input.getvalue()
  st.sidebar.image(img, width=300, height=300)
  detect = st.sidebar.button("Detect Cataract")
  dr = st.sidebar.button("Analyze Diabetic Retinopathy")
  r = st.sidebar.button("Analyze Redness Levels")
  np.set_printoptions(suppress=True)
  model = tensorflow.keras.models.load_model('/content/drive/MyDrive/ayes project/eye_models/cataract/model.h5')
  data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)
  image = Image.open(image_input)
  size = (224, 224)
  image = ImageOps.fit(image, size, Image.ANTIALIAS)
  image_array = np.asarray(image)
  normalized_image_array = (image_array.astype(np.float32) / 127.0) - 1
  data[0] = normalized_image_array
  size = st.slider("Adjust Image Size: ", 300, 1000)
  st.image(img, width=size, height=size)
  st.write("------------------------------------------------------")
  if detect:
    prediction = model.predict(data)
    class1 = prediction[0,0]
    class2 = prediction[0,1]
    
    if class1 > class2:
      st.markdown("DetAll thinks this is a **Cataract** by {:.2f}%".format(class1 * 100) )
      url1 = 'https://www.pestnet.org/fact_sheets/tomato_bacterial_spot_081.htm'
      if st.button('Guidance page '):
        st.write('the url: %s' % url1)
    elif class2 > class1:
      st.markdown("DetAll thinks this is not **Cataract** by {:.2f}%".format(class2 * 100))
    else:
      st.write("We encountered an ERROR. This should be temporary, please try again with a better quality image.")
  if dr:
    model_d = tensorflow.keras.models.load_model('/content/drive/MyDrive/ayes project/eye_models/dr/model.h5')
    data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)
    image = Image.open(image_input)
    size = (224, 224)
    image = ImageOps.fit(image, size, Image.ANTIALIAS)
    image_array = np.asarray(image)
    normalized_image_array = (image_array.astype(np.float32) / 127.0) - 1
    data[0] = normalized_image_array
    answer = model_d.predict(data)
    class1 = answer[0,0]
    class2 = answer[0,1]
    if class1 > class2:
      st.write("Diabetic Retinopathy Detected. Confidence: {:.2f}".format(class1 * 100))
      url2 = 'https://www.nei.nih.gov/learn-about-eye-health/eye-conditions-and-diseases/diabetic-retinopathy'
      if st.button('Guidance page '):
        st.write('the url: %s' % url2)
        st.write("-------------------------------")
    elif class2 > class1:
      st.write("Diabetic Retinopathy Not Detected: {:.2f}".format(class2 * 100))
      st.write("-------------------------------")
  if r:
     model_r = tensorflow.keras.models.load_model('/content/drive/MyDrive/ayes project/eye_models/redness/model.h5')
     data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)
     image = Image.open(image_input)
     size = (224, 224)
     image = ImageOps.fit(image, size, Image.ANTIALIAS)
     image_array = np.asarray(image)
     image.show()
     normalized_image_array = (image_array.astype(np.float32) / 127.0) - 1
     data[0] = normalized_image_array
     answer = model_r.predict(data)
     class1 = answer[0,0]
     class2 = answer[0,1]
     if class1 > class2:
       st.write("Redness Levels: {:.2f}%".format(class1 * 100))
       url3 = 'https://www.nei.nih.gov/learn-about-eye-health/eye-conditions-and-diseases/uveitis'
       if st.button('Guidance page '):
         st.write('the url: %s' % url3)
         st.write("-------------------------------")
     elif class2 > class1:
       st.write("No Redness Detected. Confidence: {:.2f}%".format(class2 * 100))
       st.write("-------------------------------")
```
